```python
# Sample Data
x = [1, 2, 3, 4, 5]
y = [13, 18, 11, 23, 20]
plt.plot(x, y, marker='o')
plt.title("Simple Line Chart")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
students = ["A", "B", "C", "D", "E"]
marks = [85, 70, 90, 65, 95]
plt.bar(students, marks,color=["mediumblue"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
students = ["A", "B", "C", "D", "E"]
marks = [85, 70, 90, 65, 95]
plt.bar(students, marks, color=["red", "blue", "green", "orange", "purple"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.title("Programming Language Usage")

plt.show()
```
